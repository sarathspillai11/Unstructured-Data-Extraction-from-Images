```python
import matplotlib.pyplot as plt
import keras_ocr
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python

%%html
<img src='https://www.freshbooks.com/wp-content/uploads/2018/12/pdf-PDF-Invoice3-1.png' />
```

```python
images = [keras_ocr.tools.read(images) for images in ['https://www.freshbooks.com/wp-content/uploads/2018/12/pdf-PDF-Invoice3-1.png','https://www.freshbooks.com/wp-content/uploads/2018/12/pdf-PDF-Invoice3-1.png']]
```

```python
%timeit prediction = pipeline.recognize(images)
```

```python
prediction
```

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))
print(axs)
for (ax, image, predictions) in zip(list(axs), images, prediction):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
    
```
